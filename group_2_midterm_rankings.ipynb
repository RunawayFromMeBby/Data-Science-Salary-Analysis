```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("resources/data_cleaned_2021.csv")
df
```

```python

df= df.drop(columns = ['Salary Estimate','Job Description','Headquarters','Revenue','Competitors',
                                                         'company_txt','Age','spark','aws',
                                                         'sql','sas','keras','pytorch','scikit',
                                                         'tensor','hadoop','tableau','bi','flink',
                                                         'mongo','google_an','job_title_sim', 'Size', 'Founded', 'Hourly', 'Employer provided', 'Lower Salary', 'Upper Salary', 'Python', 'excel', 'seniority_by_title', 'Degree', 'Job Title', 'index', 'Type of ownership', 'Industry', 'Sector', 'Location'])
df.drop(df[df['Rating']==-1.0].index,inplace = True)
df
```

```python
df3 = df.nlargest(5, 'Rating')
df3
```

```python
df4 = df.nsmallest(5, 'Rating')
df4
```

```python
ax = df3.plot.bar(x='Company Name', y='Rating', rot=0, figsize=(10,5))
plt.ylim(1, 5)
```

```python
ax2 = df4.plot.bar(x='Company Name', y='Rating', rot=0, figsize=(10,5))
plt.ylim(1, 5)
```
